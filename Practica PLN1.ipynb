Primero, tomaremos un sitio web y analizaremos el texto para ver de que trata la pagina web

La libreria Urlib ayuda a obtener los datos de la pagina web

```python
import urllib.request
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html= response
print (html)
```

Utilizamos Beatiful Soup que es una libreria en python para extraer datos de archivos HTML y XML. Utilizamos esta libreria para limpiar el texto de la pagina web de las etiquetas HTML.


```python
from bs4 import BeautifulSoup
soup= BeautifulSoup(html, 'html5lib')
text= soup.get_text(strip= True)
print (text)
```

Una vez que tenemos el texto plano que obtuvimos de la pagina web, vamos a convertir este texto a Tokens

```python
tokens = [t for t in text.split()]
print (tokens)
```

## Cuenta la frecuencia de palabras

NlTK ofrece una funcion freqDist() que hace el trabajo por nosotros. Ademas de esto se elimina las palabras (a, at, the, for, etc) (stop Words) ya que son necesarias para el conteo de frecuencia de palabras.


Finalmente, trazamos el grafico de las palabras que aparecen mas frecuencia en la pagina web

```python
from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens=tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
freq= nltk.FreqDist(clean_tokens)
for key, val in freq.items():
    print(str(key)+ ':'+str(val))
freq.plot(20, cumulative=False)
```
